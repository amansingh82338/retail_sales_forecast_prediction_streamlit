# Sales Forecasting: Exploratory Data Analysis (EDA)
This notebook performs EDA on the merged retail sales dataset and exports visualization outputs.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio

# Set paths
data_path = r'C:\Users\amans\Downloads\Codtech_Projects\predictive_sales_forecast\merged\merged_sales.csv'
graph_path = r'C:\Users\amans\Downloads\Codtech_Projects\predictive_sales_forecast\graphs'

# Load data
df = pd.read_csv(data_path)
df['Date'] = pd.to_datetime(df['Date'])

# Basic Overview
print(df.info())
print(df.describe())
```

```python
# 1. Weekly Sales Distribution
fig1 = px.histogram(df, x='Weekly_Sales', title='Distribution of Weekly Sales', nbins=50)
fig1.write_image(f'{graph_path}\\weekly_sales_dist.png')
fig1.show()
```

```python
# 2. Sales Trend Over Time
ts_data = df.groupby('Date')['Weekly_Sales'].sum().reset_index()
fig2 = px.line(ts_data, x='Date', y='Weekly_Sales', title='Total Weekly Sales Trend')
fig2.write_image(f'{graph_path}\\sales_trend.png')
fig2.show()
```

```python
# 3. Sales by Store
store_sales = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).reset_index()
fig3 = px.bar(store_sales, x='Store', y='Weekly_Sales', title='Total Sales by Store')
fig3.write_image(f'{graph_path}\\sales_by_store.png')
fig3.show()
```

```python
# 4. Holiday vs. Non-Holiday
fig4 = px.box(df, x='IsHoliday', y='Weekly_Sales', color='IsHoliday', title='Sales Impact: Holiday vs Non-Holiday')
fig4.write_image(f'{graph_path}\\holiday_impact.png')
fig4.show()
```

```python
# 5. Correlation Heatmap
corr = df.corr(numeric_only=True)
fig5 = px.imshow(corr, title='Correlation Matrix of Numerical Features', text_auto='.2f')
fig5.write_image(f'{graph_path}\\correlation_matrix.png')
fig5.show()
```
